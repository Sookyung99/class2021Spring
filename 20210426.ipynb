```python
%matplotlib inline
from matplotlib import pyplot as plt
import IPython.display as ipd
import numpy as np
from scipy.signal import lfilter
```

```python
def phasor(amp, sr, dur, freq):
  t = np.arange(1, sr * dur + 1) / sr
  theta = t * (2 * np.pi) * freq
  s = np.sin(theta)
  return s, theta, t
```
